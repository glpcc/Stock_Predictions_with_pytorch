```python
import torch
from torch import nn

class SMRNN(nn.Module):
    def __init__(self,input_size,outputs) -> None:
        super().__init__()
        self.input_size = input_size
        self.outputs = outputs
        self.weight_change_ln1 = nn.Linear(input_size, outputs*input_size)
        self.net3_weights = torch.zeros((input_size, outputs))
    
    def forward(self,inputs):
        self.weight_change: torch.Tensor = self.weight_change_ln1(inputs)
        self.net3_weights += torch.reshape(self.weight_change, (self.input_size,self.outputs))
        x = inputs @ self.net3_weights.t()
        return x

    def clean(self):
        self.net3_weights = torch.zeros((self.input_size, self.outputs))
```

```python
import torch.nn.functional as F
from torch import optim

net = SMRNN(3,3)
opt = optim.Adam((param for param in net.parameters() if param.requires_grad),lr=0.01)
loss_func = F.mse_loss
epochs = 2000
for i in range(epochs):
    inpt = torch.rand(3)
    y = net(inpt)
    result = 2*inpt 
    loss = loss_func(y,result)
    print(loss)
    loss.backward()
    opt.step()
    opt.zero_grad()
    net.clean()
```
